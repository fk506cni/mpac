```python
import pandas as pd
import pptx
from pptx import Presentation
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
pd.get_option("display.max_rows")
pd.set_option("display.max_rows", 1200)
```

```python
f = "../dat/wifistrength_capture/strength_pics.pptx"
```

```python
prs = Presentation(f)
```

```python
prs = Presentation(f)
l_d = []

for i, slide in enumerate(prs.slides):
    if 1 <= i <= 10:
        d_i = dict()
        d_i["title"] = slide.shapes[0].text
        d_i["text"] = slide.shapes[2].text
        l_d.append(d_i)
l_d
```

```python
for d_i in l_d:
    for tmp in d_i["text"].split():
        key, value = tmp.split(":")
        d_i[key] = value
l_d
```

```python
for d_i in l_d:
    dt = datetime.strptime(d_i["title"][11:-4], '%Y%m%d-%H%M%S')
    d_i["dt"] = dt
l_d
```

```python
wifi_df = pd.DataFrame(l_d)
wifi_df = wifi_df[['span', 'ch', 'ss', 'ls', 'dt']]
wifi_df["category"] = "wifi"
wifi_df
```

```python
f = "../dat/slack-channel-exporter_start_notify.xlsx"
send_df = pd.read_excel(f)
send_df["category"] = "send"
send_df
```

```python
f = "../dat/slack-channel-exporter_result_notify (1).xlsx"
receive_df = pd.read_excel(f)
receive_df["category"] = "receive"
receive_df
```

```python
def convert_to_datetime(date_str):
    return datetime.strptime(date_str, '%a %b %d %Y_%H:%M:%S.%f')
```

```python
send_df['dt'] = send_df['ts_str'].apply(convert_to_datetime)
send_df.head()
```

```python
receive_df['dt'] = receive_df['ts_str'].apply(convert_to_datetime)
receive_df.head()
```

```python
df = pd.concat([wifi_df, send_df, receive_df])
df = df.sort_values(by="dt").reset_index(drop=True)
df
```

```python
df[df["category"] == "wifi"]
```

「扉９０度 距離　０メートル」の前にwifiの強度を測り忘れている？

```python
df = df.iloc[111:157].reset_index(drop=True)
```

```python
wifi_index = df[df["category"] == "wifi"].index.tolist()
wifi_index.append(df.shape[0])
wifi_index
```

```python
time_diff_l = []
for i in range(len(wifi_index) - 1):
    time_diff_d = {}
    df_ = df.iloc[wifi_index[i]:wifi_index[i+1]]
    first_receive_dt = df_[df_["category"] == "receive"].iloc[0]['dt']
    last_send_dt = df_[df_["category"] == "send"].iloc[-1]['dt']
    time_diff = first_receive_dt - last_send_dt
    time_diff_d["ss"] = df_[df_["category"] == "wifi"].iloc[0]['ss']
    time_diff_d["span"] = df_[df_["category"] == "wifi"].iloc[0]['span']
    time_diff_d["time_diff"] = time_diff.total_seconds()
    time_diff_l.append(time_diff_d)
time_diff_l
```

```python
time_diff_df = pd.DataFrame(time_diff_l)
time_diff_df
```

```python
span_5200 = time_diff_df[time_diff_df['span'] == "5200"]
span_2412 = time_diff_df[time_diff_df['span'] == "2412"]

plt.figure(figsize=(10, 5))

plt.plot(span_5200['ss'], span_5200['time_diff'], marker='o', label='span = 5200')
plt.plot(span_2412['ss'], span_2412['time_diff'], marker='o', label='span = 2412')

plt.xlabel('ss')
plt.ylabel('time_diff')
plt.title('Time Difference vs SS for Different Span Values')
plt.legend()
plt.grid(True)

plt.show()
```
